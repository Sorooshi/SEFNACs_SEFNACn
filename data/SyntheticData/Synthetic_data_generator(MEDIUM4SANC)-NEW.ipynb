```python
def flat_ground_truth(ground_truth):
    k = 0
    labels_true, labels_true_indices = [], []
    for v in ground_truth:
        for vv in v:
            labels_true.append(int(k))
            labels_true_indices.append(vv)
        k += 1
        
    return labels_true, labels_true_indices
```
